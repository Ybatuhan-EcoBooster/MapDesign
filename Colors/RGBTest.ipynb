```python
#Core Libraries
import pandas as pd 

# For Colors
import webcolors
```

# Dataframes 

```python
df = pd.read_csv("colorhexa_com.csv")
df
```

```python
df_2 = pd.read_csv("wikipedia_color_names.csv")
df_2
```

```python
Colors = pd.concat([df,df_2],ignore_index=True,join="inner")
Colors
```

# Dataframe Cleaning

```python
Colors = Colors.drop_duplicates(subset=["Hex (24 bit)"])
Colors = Colors.reset_index(drop=True)
Colors["Hex"] = Colors["Hex (24 bit)"]
Colors = Colors[["Name","Hex (24 bit)","Hex","Red (8 bit)","Green (8 bit)","Blue (8 bit)"]]
Colors.columns = ["Name","Colors","Hex","R","G","B"]
Colors
```

```python
Colors["RGB"] = Colors["R"].astype(str) + "," + Colors["G"].astype(str) + "," + Colors["B"].astype(str)
Colors
```

```python
Colors["RGB"] = [tuple(map(int, item.split(","))) for item in Colors["RGB"]]
Colors
```

# Colors Style

```python
# Define a function to apply the CSS style to hex color codes
def color_styler(hex_code):
    style = f'background-color: {hex_code}; color: {hex_code};'
    return style

# Apply the CSS style to the 'Color' column
style = Colors.style.applymap(color_styler, subset=['Colors'])
style
```

# Save

```python
# Save as Excel file
style.to_excel('styled_dataframe.xlsx', index=False)

# Save as HTML file
style.to_html('styled_dataframe.html', index=False)
```
